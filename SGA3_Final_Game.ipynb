### Space Ufo 2022
In this game we have simple base model for the game for the final task.

THe game is initially has idea of "space invaders" type of game, very simplified though.

We have a spaceship here, an UFO and a magic perk which can add more energy to messiles.

The code is done with usage of Turtle to visualise.

We have a row of simple functions to define some behaviour of the objects and the game. I've added functions descriptions inline.

After the code run we need to enter numbers from 1 to 3 for difficulty level and then the game will start in the window with moving objects (our starship, UFO and perk).

Perk rendering is not very great because of Turtle behaviour and slow init and I has plan to add asyncio module to use randomized timer for perk initialization, but stuck with it, so it generates in the beginning as well.

Every UFO hit will add score to scoreboard and will render new UFO, in case UFO will reach low border - game will exit. TO exit game you can press 'q' key as described on the screen.

I have some strange behaviour with Turtle proper exit (turtle.bye()) on MacOS or Ubuntu - kernel hangs after it, but it closes OK on Windows 11 the same time. Anyway had no time to properly debug this and Turtle was not the best choice here :) Anyway this is definetily better than with no visualisation!

```python
import turtle
import time
import random

# import asyncio

# Defining move step for spaceship for X axis
spaceship_move_x = 15

# Set score to be 0 initially
score = 0

# Default speed for missile
missile_speed = 10

# Global var for main while loop for proper cycle end
running = True

#Global flag for perk function
perk_flag = True


# Function to move spaceship to the left
def move_left():
    """Function to move spaceship to the left and limit its movement to the left border"""
    x = spaceship.xcor() - spaceship_move_x
    if x < -190:
        x = -190
    spaceship.setx(x)


# Function to move spaceship to the right
def move_right():
    """Function to move spaceship to the right and limit its movement to the right border"""
    x = spaceship.xcor() + spaceship_move_x
    if x > 190:
        x = 190
    spaceship.setx(x)


# Function to fire missile
def fire_missile():
    """Changing missile position by 30 every step"""
    x = spaceship.xcor()
    y = spaceship.ycor()
    missile.setposition(x, y + 30)
    missile.showturtle()


def perk_creation():
    """Perk creation and adding speed boost for the spaceship"""
    global perk_flag
    while perk_flag:
        # time.sleep(random.randint(5, 10))
        # commented sleep here as it stops script for this period
        # tried to use asyncio here but with no luck 
        # as it was not known initially when logic was created an most code created
        global perk
        perk = turtle.Turtle()
        perk.color('yellow')
        perk.shape('circle')
        perk.up()
        perk.rt(90)
        perk.setposition(random.randint(-180, 180), 180)
        perk.speed(1)
        perk_flag = False
    return perk.fd(5)


def difficulty():
    """Difficulty change for invaders in terms of their speed"""
    inv_speed = 0
    while True:
        diff = int(input('Select difficulty from 1 to 3.'))
        if diff == 1:
            inv_speed = 2
            break
        elif diff == 2:
            inv_speed = 4
            break
        elif diff == 3:
            inv_speed = 6
            break
        else:
            print('Select difficulty from 1 to 3.')
    return inv_speed


# Function for proper quitting using 'q' key
def quit_game():
    """To quit turtle properly without kernel error when using 'while True' for main section"""
    global running
    running = False


# Window init
win = turtle.Screen()
win.setup(width=500, height=500)
win.bgcolor('midnightblue')
win.title("Space ufo 2022")

# Create square border for the game
border = turtle.Turtle()
border.speed(0)
border.color('white')
border.up()
border.setposition(-200, -200)
border.down()
border.pensize(3)
for side in range(4):
    border.fd(400)
    border.lt(90)  # Turn left 90 degree
border.hideturtle()

# Score board
score_marker = turtle.Turtle()
score_marker.speed(0)
score_marker.color('white')
score_marker.up()
score_marker.setposition(-200, 210)
score_marker.write('Score: %s. For exit press "q".' % score)
score_marker.hideturtle()

# Spaceship creation
spaceship = turtle.Turtle()
spaceship.shape('triangle')
spaceship.color('green')
spaceship.up()
spaceship.speed(0)
spaceship.setposition(0, -180)
spaceship.setheading(90)

# Create spaceship's missile​
missile = turtle.Turtle()
missile.color('yellow')
missile.shape('triangle')
missile.up()
missile.speed(0)
missile.setheading(90)
missile.shapesize(0.5, 0.5)
missile.hideturtle()

# Ufo creation
ufo = turtle.Turtle()
ufo.color('red')
ufo.shape('square')
ufo.up()
ufo.speed(0)
ufo.setposition(-180, 180)

# Keyboard binding for keyboard keys
turtle.listen()
turtle.onkey(move_left, 'Left')
turtle.onkey(move_right, 'Right')
turtle.onkey(fire_missile, 'space')
turtle.onkeypress(quit, 'q')

#=================================================================
# Main routine
#=================================================================

# Init ufo difficulty
ufo_speed = difficulty()

while running:

    # Check for the border and move step down for ufo
    if ufo.xcor() > 190:
        ufo.rt(90)
        ufo.fd(30)
        ufo.rt(90)
    elif ufo.xcor() < -190:
        ufo.lt(90)
        ufo.fd(30)
        ufo.lt(90)
    elif ufo.ycor() < -165:
        win.bye()
        # TODO game over
    ufo.fd(ufo_speed)

    # Perk for missile
    perk_creation()

    # Fire the missile
    missile.fd(missile_speed)

    # Check for hit
    if abs(missile.xcor() - ufo.xcor()) < 15 and abs(missile.ycor() - ufo.ycor()) < 15:
        # Update the score
        score = score + 1
        score_marker.clear()
        score_marker.write('Score: %s. For exit press "q".' % score)

        # Reset ufo
        missile.hideturtle()
        ufo.hideturtle()
        time.sleep(1)
        ufo.showturtle()
        ufo.setposition(random.randint(-180, 180), 180)

        # Check for perk collection
    if abs(spaceship.xcor() - perk.xcor()) < 15 and abs(spaceship.ycor() - perk.ycor()) < 15:
        #Modify missile speed
        missile_speed += 5
        perk.hideturtle()

turtle.bye()
```
